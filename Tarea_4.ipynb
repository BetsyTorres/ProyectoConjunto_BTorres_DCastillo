**Ejercicio 2:** Generación variable aleatoria Discreta

> La distribución binomial modela el número de éxitos de n ensayos independientes donde hay una probabilidad p de éxito en cada ensayo.
>Generar una variable aletoria binomial con parámetros $n=10$ y $p=0.7$. Recordar que $$X\sim binomial(n,p) \longrightarrow p_i=P(X=i)=\frac{n!}{i!(n-i)!}p^i(1-p)^{n-i},\quad i=0,1,\cdots,n$$
> Por propiedades de la operación factorial la anterior $p_i$ se puede escribir como:
> $$p_{i+1}=\frac{n-i}{i+1}\frac{p}{1-p} p_i $$

### <mark  style="background-color: lavender"> **Sol 2 Betsy Torres** </mark>
$$p_{i+1}=P(X=i+1)=\frac{n!}{i+1!(n-(i+1))!}p^{i+1}(1-p)^{n-i-1}$$

considerando la siguiente propiedad: $ p^{i+1}=p^ip^1$

Aplicamos la propiedad: $ \frac{p_{i+1}}{p_i}=\frac{n-1}{i+1}\frac{p}{1-p}$

Finalmente obtenemos lo siguiente:

$$ p_{i+1}=\frac{n-i}{i+1}\frac{p}{1-p}p_i$$


```python
# Sol 2 Betsy Torres
# Defino la función para generar variables aleatorias discretas
def Gen_distr_discreta(U:'vector de números aleatorios',
                       p_acum: 'P.Acumulada de la distribución a generar'):
    '''Tener en cuenta que este arreglo cuenta números empezando del 0'''
    v = np.array(list(map(lambda j:sum(1 for i in p_acum if i<U[j]),range(len(U)))))
    return v

def plot_histogram_discrete(distribucion:'señal de varibles aleatorias de un distribución DISCRETA dada',
                            label:'label del legend a aparecer en el gráfica',
                            densidad:'por defecto regresa el histograma en densidad'=True):
    # len(set(distribucion)) cuenta la cantidad de elementos distintos de la variable 'distribucion'
    plt.figure(figsize=[8,4])
    y,x = np.histogram(distribucion,bins = len(set(distribucion)),density = densidad) 
    plt.bar(x[1:],y,label=label)
#     print(len(y),len(x))
    plt.legend()
    plt.show()
    
# Función que calcula la probabilidad acumulada optimizada
def P_acum_Binomial_o(n:'Numero de repeticiones',p:'Probabilidad de éxito'):
    '''
    Esta función genera la funciín de distribución de probabilidad acumulada de una distribución binomial.
    '''
    Pr = np.zeros(n)
    Pr[0] = (1-p)**n
    def pr(i):
        nonlocal Pr
        c  = p/(1-p)
        Pr[i+1]=(c*(n-i)/(i+1))*Pr[i]
        
    # Lleno el vector Pr usando compresión de listas
    [pr(i) for i in range(n-1)]
    return np.cumsum(Pr)
    

# Variables iniciales para la distribución binomial
n, p, N = 10, 0.7, 10**4

p_bin = P_acum_Binomial_o(n,p)
u = np.random.rand(N)


# Empleando el método de la transformada inversa
dist_binomial_disc = Gen_distr_discreta(u,p_bin)
plot_histogram_discrete(dist_binomial_disc,'Variable aleatoria binomial acumulada')
```

```python
# Sol 2 Daniel Castillo

# Defino la función para generar variables aleatorias discretas
def f_discreta(u, p_acum):
    a = np.array(list(map(lambda i:sum(1 for j in p_acum if j<u[i]),range(len(u)))))
    return a

def graph(dist,label,density=True):
    plt.figure(figsize=[8,4])
    y,x = np.histogram(dist,bins = len(set(dist)),density = density) 
    plt.bar(x[1:],y,label=label)
    plt.legend()
    plt.show()
    
def f_acum(n,p):
    v = np.zeros(n)
    v[0] = (1-p)**n
    def aux(i):
        nonlocal v
        c  = p/(1-p)
        v[i+1]=(c*(n-i)/(i+1))*v[i]      
    [aux(i) for i in range(n-1)]
    return np.cumsum(v)
    
N = 10000
n = 10
p = 0.7


f_bin = f_acum(n,p)
u = np.random.rand(N)

dist_binomial_disc = f_discreta(u,f_bin)
graph(dist_binomial_disc,'Variable aleatoria binomial acumulada')
```

```python
def Acep_rechazo_x(f:'función objetivo a generar',
                   t:'función que mayora a f', 
                   N: 'Variables aleatorioas a genrar'):
    '''Esta función es para, con base en la función de aceptación y rechazo, 
    poder tener N variables que son aceptadas por el método'''
    aceptados = [x for x in range(N)] #generar vector de puntos
    def llenar(i):
        nonlocal f, t, aceptados
        aleatorio_1=np.random.rand() #valor aleatorio 1
        aleatorio_2=np.random.rand() #valor aleatorio 2
        f_x = f(aleatorio_1) #evaluar valor aleatorio en la pdf
        t_x = t #funcion constante
        condition = aleatorio_2*t_x <= f_x #validar condición 
        while condition==False: #repetir si la condición es falsa
            aleatorio_1 = np.random.rand()
            aleatorio_2 = np.random.rand()
            f_x = f(aleatorio_1)
            t_x = t
            condition = aleatorio_2*t_x <= f_x 

        aceptados[i] = aleatorio_1 #añadir cuando la condición se haga verdadera
    [llenar(i) for i in range(N)]
    return aceptados


# Ilustración del método de aceptación y rechazo cuando se toma t(x) constante

# Función objetivo 
f3 = lambda x: 20*x*((1-x)**3)
# Máximo de la función f
max_f3 = f3(optimize.fmin(lambda x:-f3(x),0,disp=False))
# Función t -> Función constante
t3 = max_f3 # función constante

x33 = np.arange(0,1,0.01)  # Rango donde se graficará las funciones
print('El máximo de f es:',max_f3)

N33 = 10000 # número de puntos a simular

v_a_generada = Acep_rechazo_x(f3,t3,N33)
plt.hist(v_a_generada, density=True, bins=40, label="Variable aleatoria generada")
plt.plot(x33, f3(x33), label="Función de distribución de probabilidad")
plt.legend()
plt.show()
```

### Ejercicio Poison     
Seguir un procedimiento similar al mostrado cuando se generó una distribución binomial, pero en esta caso genere un código que genere variables aletorias Poisson cuya función de distribución de probabilidad esta dada por:
$$P(k,\lambda)=\frac{e^{-\lambda}(\lambda)^k}{k!}$$
    
> Demuestre matemáticamente que 
$$P(k+1)=\frac{\lambda}{k+1}P(k)$$
y a partir de esta relación genere variables aletorias que distribuyen poisson usando el método de la transformada inversa.

Enlace: https://es.wikipedia.org/wiki/Distribuci%C3%B3n_de_Poisson

### <mark  style="background-color: lavender">**Sol Poison Betsy Torres** </mark>



Primero desarrollo $p(k+1)$ siguiendo la fórmula de $p(k)$ obteniendo lo siguiente:

$$p(k+1)=\frac{\exp^{-\lambda} \lambda^{k+1}{k+1}}{(k+1)!}$$

Después de dividr el término entre $p(k)$ y desarrollamos el exponente y factorial, simplificandolo obtendremos lo siguiente: 

$$\frac{p(k+1)}{p(k)}=\frac{\lambda}{k+1}$$

Ahora despejamos $p(k+1)$ y obtendremos la ecuación que estabamos buscando: 
$$p(k+1)=\frac{\lambda}{k+1}p(k)$$

```python
def P_acum_Poisson(tasa: 'Igual al parámetro lambda que indica la media o la tasa de ocurrencia del evento',
                   k: 'Número de ocurrencias del evento'):
    Pr = np.zeros(k)
    Pr[0] =np.exp(-tasa)
    def pr(i):
        nonlocal Pr
        Pr[i+1]=(tasa*Pr[i])/(i+1)
        
    [pr(i) for i in range(k-1)]
    return np.cumsum(Pr)

#ya que definimos la función de prob acum poison aplicamos podemos aplicar método de la transformada inversa

lam, k, N=4, 10, 10**4

p_acum_p = P_acum_Poisson(lam,k)
u = np.random.rand(N)


d_poisson = Gen_distr_discreta(u,p_acum_p)
plot_histogram_discrete(d_poisson,'usando método de transformada inversa') #aplicando el metodo

d_poisson_numpy = np.random.poisson(lam,N)
plot_histogram_discrete(d_poisson_numpy,'usando función de numpy binomial') #aplicando funciones de numpy
```
